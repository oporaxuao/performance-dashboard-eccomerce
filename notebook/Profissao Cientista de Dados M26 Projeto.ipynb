Abaixo temos a configuração do ambiente SQL:

```python
import sqlite3
import pandas as pd
```

```python
# Lendo os arquivos
df_clientes = pd.read_csv('TB_CLIENTES_PROJETO_ECOMM.csv', sep=';')
df_transacoes = pd.read_csv('TB_TRANSACOES_PROJETO_ECOMM.csv', sep=';')

# Tratando o preço para não dar erro no BI depois
df_transacoes['Price'] = df_transacoes['Price'].str.replace(',', '.').astype(float)

# Criando a conexão e as tabelas
conn = sqlite3.connect(':memory:')

# Vamos usar nomes em MAIÚSCULO para as tabelas
df_clientes.to_sql('TB_CLIENTES', conn, index=False, if_exists='replace')
df_transacoes.to_sql('TB_TRANSACOES', conn, index=False, if_exists='replace')
```

```python
conn = sqlite3.connect('projeto.db')
# Carregar o DataFrame no banco de dados SQLite - criando tb_transacoes e tb_clientes
df_transacoes.to_sql('tb_transacoes', conn, index=False, if_exists='replace')
df_clientes.to_sql('tb_clientes', conn, index=False, if_exists='replace')
```

```python
# Função para executar consultas SQL e retornar o resultado como um DataFrame
def run_query(query):
    return pd.read_sql_query(query, conn)
```

```python
query = """
SELECT
    t.id_client,
    t.Category,
    t.Price,
    t."Card Type",
    c.First_name,
    c.Gender,
    c.Job_Title,
    c.state_name
FROM TB_TRANSACOES AS t
INNER JOIN TB_CLIENTES AS c ON t.id_client = c.Id_client
"""
result_df = run_query(query)
print(result_df)
```

Justifique a escolha do JOIN:

utilizei o INNER JOIN para consolidar as informações das duas tabelas, pois meu objetivo era obter apenas os registros de transações que possuem correspondentes na tabela de clientes. O INNER JOIN retorna apenas as linhas onde há uma correspondência em ambas as tabelas, garantindo que os dados consolidados sejam relevantes e completos para análise. Se eu tivesse utilizado um LEFT JOIN, por exemplo, poderia ter incluído transações sem informações de clientes, o que não seria útil para a análise pretendida.

Exportando o arquivo como CSV:

```python
result_df.to_csv('dados_ecommerce_final.csv', index=False)
```
